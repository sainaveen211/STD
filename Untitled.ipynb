```python
import pandas as pd

# Load the dataset
data = pd.read_csv("Used Car Dataset.csv")

# Display the first few rows of the dataset
print(data.head())

# Summary statistics of the numerical variables
print(data.describe())

# Check for missing values
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Plotting a histogram of car prices
plt.figure(figsize=(10, 6))
plt.hist(data['insurance_validity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of insurance_validity')
plt.xlabel('insurance_validity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting a scatter plot of mileage vs. price
plt.figure(figsize=(10, 6))
plt.scatter(data['mileage(kmpl)'], data['price(in lakhs)'], color='green', alpha=0.5)
plt.title('Mileage vs. Price')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python
import seaborn as sns

# Calculating the correlation matrix
correlation_matrix = data.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
